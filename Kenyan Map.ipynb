```python
lat=list(ke_df['lat'])
long=list(ke_df['lng'])
city=list(ke_df['city'])
pop=list(ke_df['population'])
```

```python
def population_difference(population):
    if population <100000:
        return 'green'
    elif population > 200000 <= 500000:
        return 'blue'
    else:
        return 'red'

fg=fm.FeatureGroup(name='My_map')
my_map=fm.Map([0.0236, 36.9062],zoom_start=8,tiles='Stamen Terrain')
my_map.add_child(fg)
for lt,lg,pin,popul in zip(lat,long,city,pop):
    fg.add_child(fm.Marker(location=[lt,lg],popup=str(pin)+'  '+str(popul),icon=fm.Icon(color=population_difference(popul)))).add_to(my_map)
my_map
```

```python
def population_difference(population):
    if population <100000:
        return 'green'
    elif population > 200000 <= 500000:
        return 'blue'
    else:
        return 'red'

fg=fm.FeatureGroup(name='My_map')
my_map=fm.Map([0.0236, 36.9062],zoom_start=8,tiles='Stamen Terrain')
my_map.add_child(fg)
for lt,lg,pin,popul in zip(lat,long,city,pop):
    fg.add_child(fm.CircleMarker(location=[lt,lg],popup=str(pin)+'  '+str(popul),radius=7,fill=True,fill_opacity=0.7,fill_color=population_difference(popul))) .add_to(my_map)
my_map
```

```python
import geojson
with open('kenya geo.geojson') as f:
    data = geojson.load(f)
```

```python
fg=fm.FeatureGroup(name='My_map')

My_map=fm.Map([0.0236, 36.9062],zoom_start=7,zoom_control=7,tiles='Stamen Terrain')
My_map
# Inserting geo json map to mark boundaries
fg.add_child(fm.Choropleth(geo_data=data,
                            name='Chloropleth',
                            data=ke_df,
                            columns=['city','population'],
                            key_on= "id",
                            fill_color='BrBG',
                            fill_opacity=0.8,
                            line_opacity=0.4,
                            legend_name='Pop_per_city').add_to(My_map))
fm.LayerControl().add_to(My_map)
```
